```python
import numpy as np  
import matplotlib.pyplot as plt
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
# Note: Also add title, xlabel, and ylabel to the plot.

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.title('Assignment')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.scatter(x=x,y=y)
plt.show()
# A scatter plot is a graphical representation of the relationship between two variables. It is a type of plot that uses Cartesian coordinates to display values for two variables for a set of data points. The data points are represented as individual points on the plot, with the horizontal axis(x-axis) representing one variable and the vertical axis(y-axis) representing the other variable.

# Each data point on the scatter plot represents the values of both variables for a single observation or measurement. By examining the distribution of points on the plot, you can visually assess the strength and direction of the relationship between the two variables.
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:

x1,y1= np.array([0, 1, 2, 3, 4, 5]) , np.array([0, 100, 200, 300, 400, 500])
x2,y2= np.array([0, 1, 2, 3, 4, 5]) , np.array([50, 20, 40, 20, 60, 70])
x3,y3= np.array([0, 1, 2, 3, 4, 5]) , np.array([10, 20, 30, 40, 50, 60])
x4,y4= np.array([0, 1, 2, 3, 4, 5]) , np.array([200, 350, 250, 550, 450, 150])
# The subplot() function is used to create multiple plots in a single figure. It allows you to arrange multiple plots in a grid pattern, with each plot sharing the same figure and axes.

# Create figure and subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Create line plots for each subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Set titles for each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Show the plot
plt.show()

```

```python

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

fig,axs=plt.subplots(nrows=1,ncols=2,figsize=(10,5))

axs[0].bar(company, profit)
axs[0].set_title('Bar Plot')
axs[1].barh(company, profit)
axs[1].set_title('Horizontal Bar Plot')

plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')

plt.show()
```
