# Q1. What is Flask Framework? What are the advantages of Flask Framework? 

```python
# Flask is a micro web framework for Python that is designed to be lightweight, flexible, and easy to use. It is ideal for building small to medium-sized web applications, APIs, and microservices.

# The advantages of Flask framework are:

# Lightweight: Flask is a micro-framework, which means it is lightweight and has a small codebase. This makes it easier to learn, use and deploy.

# Flexible: Flask is highly flexible, allowing developers to easily customize and extend the framework to meet their specific needs.

# Easy to use: Flask has a simple and intuitive interface that makes it easy for developers to build web applications quickly.

# Python-based: Flask is built in Python, which is a popular and widely used programming language. This means that developers can leverage their existing Python knowledge and skills to build web applications with Flask.

# Extensive documentation: Flask has extensive documentation that makes it easy for developers to learn and use the framework.

# Modular design: Flask has a modular design that allows developers to choose and use only the components they need, making it highly scalable.

# Large community: Flask has a large and active community of developers who contribute to its development, provide support, and share knowledge.
```

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
# url_for() function is used to build URLs for Flask routes. This function takes the name of the Flask function as its first argument and any additional keyword arguments represent the variable parts of the URL. Flask uses the function name and keyword arguments to generate the URL that matches the pattern specified in the route.

from flask import Flask, url_for

app = Flask(__name__)


@app.route('/')
def index():
    return 'This is the index page.'


@app.route('/user/<username>')
def user_profile(username):
    return f'This is the profile page for {username}.'


with app.test_request_context():
    print(url_for('index'))
    print(url_for('user_profile', username='john'))
```
